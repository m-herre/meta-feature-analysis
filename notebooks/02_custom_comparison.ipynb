# Custom Comparison

Example orchestration notebook for swapping in a different pairwise comparison.

```python
from __future__ import annotations

import sys
from pathlib import Path

notebook_dir = Path.cwd()
project_dir = notebook_dir.parent
repo_root = project_dir.parent

sys.path.insert(0, str(project_dir / "src"))
sys.path.insert(0, str(repo_root / "tabarena" / "tabarena"))

import yaml

from mfa.config import parse_config
from mfa.pipeline import run_analysis
```

```python
raw_config = yaml.safe_load((project_dir / "configs" / "default.yaml").read_text())
raw_config["comparisons"] = [
    {
        "name": "tree_vs_nn",
        "group_a": "tree",
        "group_b": "nn",
        "expected_direction": "positive",
    }
]

config = parse_config(raw_config)
result = run_analysis(config, datasets=["diabetes", "credit-g"])
result.correlation_results[:3]
```
