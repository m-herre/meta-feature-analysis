```python
# --- pymfe smoke test on real TabArena data (small dataset, all splits) ---
import time
import traceback

import numpy as np
import openml
import pandas as pd
from pymfe.mfe import MFE
from tabarena.benchmark.task.openml import OpenMLTaskWrapper
from tabarena.nips2025_utils.fetch_metadata import load_task_metadata

from mfa.metafeatures.pymfe_features import extract_pymfe_features

# Pick the smallest dataset by training samples
metadata = load_task_metadata()
smallest = metadata.sort_values("n_samples_train_per_fold").iloc[0]
dataset_name = smallest["name"]
tid = int(smallest["tid"])
print(
    f"Dataset: {dataset_name}  (tid={tid}, ~{int(smallest['n_samples_train_per_fold'])} train samples/fold)"
)

task = OpenMLTaskWrapper(openml.tasks.get_task(tid))

# TabArena uses 3 repeats × 3 folds = 9 splits
N_REPEATS = 3
N_FOLDS = 3
PYMFE_GROUPS = ("general", "statistical", "info-theory")
PYMFE_SUMMARY = ("mean", "sd")

results = []
errors = []

for repeat in range(N_REPEATS):
    for fold in range(N_FOLDS):
        label = f"r{repeat}/f{fold}"
        try:
            t0 = time.perf_counter()
            X_train, y_train, X_test, y_test = task.get_train_test_split(
                fold=fold, repeat=repeat
            )
            feats = extract_pymfe_features(
                X_train, y_train, groups=PYMFE_GROUPS, summary=PYMFE_SUMMARY
            )
            elapsed = time.perf_counter() - t0
            n_feats = len(feats)
            n_nan = sum(
                1
                for v in feats.values()
                if v is None or (isinstance(v, float) and np.isnan(v))
            )
            results.append(
                {
                    "split": label,
                    "n_features": n_feats,
                    "n_nan": n_nan,
                    "time_s": round(elapsed, 2),
                }
            )
            print(
                f"  {label}: {n_feats} features extracted ({n_nan} NaN) in {elapsed:.1f}s"
            )
        except Exception:
            errors.append({"split": label, "error": traceback.format_exc()})
            print(f"  {label}: FAILED")
            traceback.print_exc()

print(f"\n--- Summary ---")
print(f"Splits OK : {len(results)}/{N_REPEATS * N_FOLDS}")
print(f"Splits ERR: {len(errors)}/{N_REPEATS * N_FOLDS}")

if results:
    df_results = pd.DataFrame(results)
    print(
        f"\nFeatures per split: min={df_results.n_features.min()}, max={df_results.n_features.max()}"
    )
    print(
        f"NaN per split:      min={df_results.n_nan.min()}, max={df_results.n_nan.max()}"
    )
    print(
        f"Time per split:     min={df_results.time_s.min()}s, max={df_results.time_s.max()}s"
    )

    # Check feature names are consistent across splits
    if len(set(df_results.n_features)) == 1:
        print("✓ Feature count consistent across all splits")
    else:
        print("✗ Feature count varies across splits!")

    display(df_results)

if errors:
    print("\n--- Errors ---")
    for e in errors:
        print(f"\n{e['split']}:\n{e['error']}")
```
